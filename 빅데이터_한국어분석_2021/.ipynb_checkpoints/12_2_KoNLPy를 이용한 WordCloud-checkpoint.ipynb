## 한글 자연어 처리 기초

* http://konlpy.org/ko/v0.5.1/ 

```python
from konlpy.tag import Okt, Komoran, Hannanum, Kkma
```

```python
def get_tokenizer(get_tokenizer_name):
    if get_tokenizer_name == "Okt":
        tokenizer=Okt()
    elif get_tokenizer_name == "komoran":
        tokenizer=Komoran()
    elif get_tokenizer_name == "Hannanum":
        tokenizer=Hannanum()
    else:
        tokenizer=Kkma()
    return tokenizer
```

```python
from wordcloud import WordCloud, STOPWORDS

import numpy as np
from PIL import Image
```

```python
text = open('alice.txt').read()
alice_mask = np.array(Image.open('alice_mask.png'))

stopwords = set(STOPWORDS)
stopwords.add("said")
```

### 예제2: a_new_hope.txt 

```python
text = open('a_new_hope.txt').read()

text = text.replace('HAN', 'Han')
text = text.replace("LUKE'S", 'Luke')

mask = np.array(Image.open('stormtrooper_mask.png'))
```

```python
stormtrooper_mask = np.array(Image.open('stormtrooper_mask.png'))

plt.figure(figsize=(8,8))
plt.imshow(stormtrooper_mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
stopwords = set(STOPWORDS)
stopwords.add("int")
stopwords.add("ext")
```

```python
wc = WordCloud(max_words=1000, mask=mask, stopwords=stopwords, 
               margin=10, random_state=1).generate(text)

default_colors = wc.to_array()
```

```python
import random
def grey_color_func(word, font_size, position, orientation, 
                    random_state=None, **kwargs):
    return 'hsl(0, 0%%, %d%%)' % random.randint(60,100)
```

```python
plt.figure(figsize=(12,12))
plt.imshow(wc.recolor(color_func=grey_color_func, random_state=3),
          interpolation='bilinear')
plt.axis('off')
plt.show()
```

### 예제3: 한글 Word Cloud
- 육아휴직관련 법안 대한민국 국회 제 1809890호 의안

```python
import nltk
```

```python
from konlpy.corpus import kobill
files_ko = kobill.fileids()
doc_ko = kobill.open('1809890.txt').read()
```

```python
doc_ko
```

```python
from konlpy.tag import Twitter
tokenizer = get_tokenizer("Okt")
tokens_ko = tokenizer.nouns(doc_ko)
tokens_ko
```

```python
ko = nltk.Text(tokens_ko, name='대한민국 국회 의안 제 1809890호')
ko
```

```python
print(len(ko.tokens))        # returns number of tokens (document length)
print(len(set(ko.tokens)))   # returns number of unique tokens
ko.vocab()                   # returns frequency distribution
```

```python
plt.figure(figsize=(12,6))
ko.plot(50)
plt.show()
```

```python
stop_words = ['.', '(', ')', ',', "'", '%', '-', 'X', ').', '×','의','자','에','안','번',
                      '호','을','이','다','만','로','가','를']

ko = [each_word for each_word in ko if each_word not in stop_words]
ko
```

```python
ko = nltk.Text(ko, name='대한민국 국회 의안 제 1809890호')

plt.figure(figsize=(12,6))
ko.plot(50)     # Plot sorted frequency of top 50 tokens
plt.show()
```

```python
dir(ko)
```

```python
ko.count('초등학교')
```

```python
plt.figure(figsize=(12,6))
ko.dispersion_plot(['육아휴직', '초등학교', '공무원'])
```

```python
ko.concordance('초등학교')
```

```python
ko.collocation_list(4)
```

```python
mask = np.array(Image.open('brain.png'))

from wordcloud import ImageColorGenerator

image_colors = ImageColorGenerator(mask)
```

```python
data = ko.vocab().most_common(150)

# for win : font_path='c:/Windows/Fonts/malgun.ttf'
wordcloud = WordCloud(font_path='c:/Windows/Fonts/malgun.ttf',
                      relative_scaling = 0.2,
                      mask=mask,
                      background_color='white',
                      min_font_size=1,
                      max_font_size=100
                      ).generate_from_frequencies(dict(data))
plt.figure(figsize=(12,8))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```
