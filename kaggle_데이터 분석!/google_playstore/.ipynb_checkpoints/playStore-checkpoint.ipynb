```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline 

from matplotlib import font_manager, rc
plt.rcParams['axes.unicode_minus'] = False
f_path = "C:/Windows/Fonts/malgun.ttf"
font_name=font_manager.FontProperties(fname=f_path).get_name()
rc('font', family=font_name)
```

```python
# 데이터 불러오기

gps = pd.read_csv('./googleplaystore.csv')
gps.head()
```

```python
# 데이터 정보 확인

gps.info()
```

```python
gps.describe()
```

```python
gps.describe(include = np.object_)
```

```python
# 결측치 확인

gps.isnull().sum()
```

```python
# 중복 데이터 확인 및 삭제

gps.duplicated().sum()
```

```python
gps.drop_duplicates(inplace = True)
```

```python
# Category 열

gps['Category'].unique()
```

```python
gps[gps['Category'] == '1.9']
```

```python
gps.drop(gps.loc[gps['Category'] == '1.9'].index, inplace = True)
```

```python
# 결측치 처리

gps.dropna(subset = ['Rating'], inplace = True)
```

```python
# 자료형 변환

gps['Reviews'] = pd.to_numeric(gps['Reviews'])
```

```python
gps['Size'] = gps['Size'].str.replace('Varies with device','0')
gps['Size'] = (gps['Size'].replace(r'[kM]+$', '', regex=True).astype(float) *\
               gps['Size'].str.extract(r'[\d\.]+([kM]+)', expand=False).fillna(1).replace(['k','M'], [10**3, 10**6]).astype(int))
```

```python
gps['Installs'] = gps['Installs'].str.replace('+','')
gps['Installs'] = gps['Installs'].str.replace(',','')
gps['Installs'] = pd.to_numeric(gps['Installs'])
```

```python
gps['Price'] = gps['Price'].str.replace('$','').astype('float')
```

```python
gps.info()
```

```python
# 카테고리별 앱의 개수

sns.set(rc = {'figure.figsize':(20,12)}, font_scale=1.5, style = 'whitegrid')
apps = gps['Category'].value_counts()
sns.barplot(x = apps.values, y =apps.index)
plt.show()
```

```python
gps['Category'].value_counts(normalize = True) *100
```

<span style="font-size:16px ; font-style:bold;"> ==> Family, Game, Tools, Productivity, Finance에 관한 앱이 50% 정도의 비율을 차지함</span>

```python
# 사용가능 연령별 앱의 개수

apps = gps['Content Rating'].value_counts()
sns.barplot(x = apps.index, y = apps.values)
plt.show()
```

<span style="font-size:16px ; font-style:bold;"> ==> 특정 연령을 대상으로 하는 앱보다는 불특정 다수를 대상으로 하는 앱이 많은 비율을 차지함</span>

```python
# 가격대별 앱의 개수

gps['PriceRange'] = 0
gps['PriceRange'] = pd.cut(gps.Price, bins = [-1, 0, 10, 100, 1000], 
                           labels = ['Free','$10 or less','$100 or less','more than $100'])
```

```python
apps = gps['PriceRange'].value_counts()

colors = ['#ff9999', '#ffc000', '#8fd9b6', '#d395d0']
wedgeprops={'width': 0.6, 'edgecolor': 'w', 'linewidth': 0.5}

plt.pie(apps.values, labels=apps.index, autopct='%.1f%%', startangle=0, counterclock=False, 
        colors=colors, wedgeprops=wedgeprops, explode = (0,0,0,0.4))
plt.show()
```

<span style="font-size:16px ; font-style:bold;"> ==> 대부분이 무료 앱 </span>

```python
# 카레고리별 Rating

sns.set(rc = {'figure.figsize':(20,10)}, font_scale=1.5, style = 'whitegrid')
chart = sns.boxplot(x = 'Category', y = 'Rating', data = gps)
chart.set_xticklabels(chart.get_xticklabels(),rotation = 90)
plt.show()
```

```python
# 카레고리별 Installs

sns.set(rc = {'figure.figsize':(20,10)}, font_scale=1.5, style = 'whitegrid')
chart = sns.boxplot(x = 'Category', y = np.log10(gps['Installs']), data = gps)
chart.set_xticklabels(chart.get_xticklabels(),rotation = 90)
plt.show()
```

```python
# Rating과 Installs의 관계

sns.scatterplot( x = np.log10(gps['Installs']), y= 'Rating', data = gps, hue = 'Type', s = 100)
plt.show()
```

<span style="font-size:16px ; font-style:bold;"> ==> Install의 수가 많은 앱일수록 평점(Rating)이 좋은 편임. 유료앱보다는 무료앱의 경우 평점이 좋음 </span>

```python
# Rating과 Reviews의 관계

sns.scatterplot(x = 'Rating', y = np.log10(gps['Reviews']), data = gps, hue = 'Type', s = 100)
plt.show()
```

<span style="font-size:16px ; font-style:bold;"> ==> 평점(Rating)이 좋은 앱일수록 이용자들이 사용 후기(Review)를 작성하는 경향이 있음 </span>

```python
# Installs 된 앱 중 상위 50개

top50 = gps.sort_values(by = 'Installs', ascending = False)[:50]
top50.head()
```

```python
# Install 된 앱 중 상위 50의 카테고리

top50['Category'].value_counts(normalize = True) *100
```

```python
apps = top50['Category'].value_counts()
wedgeprops={'width': 0.6, 'edgecolor': 'w', 'linewidth': 0.5}

plt.pie(apps.values, labels=apps.index, autopct='%.1f%%', 
        startangle=0, counterclock=False, wedgeprops=wedgeprops)
plt.show()
```

<span style="font-size:16px ; font-style:bold;"> ==> 구글 앱 스토어에서 점유율이 높은 카테고리중 Install된 수가 많은 것은 Game 관련 앱임<br>
==> Family는 높은 점유율에 비해 상대적으로 Install이 적게 됨 <br>
==> Tools나 Productivity 관련 앱은 Install된 비율이 점유율과 비슷하게 차지함 </span>

```python
# Install 된 앱 중 상위 50개의 카테고리별 Rating 평균

round(top50.groupby('Category')['Rating'].mean(),2)
```

<span style="font-size:16px ; font-style:bold;"> ==> Photography 와 Game 관련 앱이 평점(Rating)이 좋음 <br>
==> 대부분의 앱이 4.0이상의 평가를 받음 </span>

```python
# Install 된 수가 많고 Rating 이 높은 GAME 앱

game = gps[gps['Category'] == 'GAME']
game = round(game.groupby('Type')[['Rating','Installs']].mean(),2)
game
```

<span style="font-size:16px ; font-style:bold;"> ==> Game 관련 앱의 경우 유료앱이 무료앱보다 평균 1.0 정도 더 좋은 평가를 받음 </span>
