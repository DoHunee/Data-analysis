### Axes 객체와 subplot 명령
- 하나의 윈도우(Figure)안에 여러개의 플롯을 배열 형태로 보이게 함
- http://matplotlib.org/api/axes_api.html#matplotlib.axes.Axes
- subplot 명령을 사용하여 명시적으로 Axes 객체를 얻음
- tight_layout 명령을 실행하면 플롯간의 간격을 자동 조정

```python
# 2 x 1의 plot
plt.subplot(1, 2, 1)
# 여기에서 윗부분에 그릴 플롯 명령 실행
plt.subplot(1, 2, 2)
plt.tight_layout()
# 여기에서 아랫부분에 그릴 플롯 명령 실행
```

```python
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

ax1 = plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'yo-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')

ax2 = plt.subplot(2, 1, 2)
plt.plot(x2, y2, 'r.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.tight_layout()
plt.show()
```

```python
np.random.seed(0)

plt.subplot(221)    # (2, 2, 1) 상, 좌
plt.plot(np.random.rand(5))
plt.title("axes 1")

plt.subplot(222)    # (2, 2, 2) 상, 우
plt.plot(np.random.rand(5))
plt.title("axes 2")

plt.subplot(223)    # (2, 2, 3) 하, 좌
plt.plot(np.random.rand(5))
plt.title("axes 3")

plt.subplot(224)    # (2, 2, 4) 하, 우
plt.plot(np.random.rand(5))
plt.title("axes 4")

plt.tight_layout()
plt.show()
```

```python
# subplots 명령으로 복수의 Axes 객체를 동시에 생성
fig, axes = plt.subplots(3, 2)

np.random.seed(0)
axes[0, 0].plot(np.random.rand(5))
axes[0, 0].set_title("axes 1")
axes[0, 1].plot(np.random.rand(5))
axes[0, 1].set_title("axes 2")
axes[1, 0].plot(np.random.rand(5))
axes[1, 0].set_title("axes 3")
axes[1, 1].plot(np.random.rand(5))
axes[1, 1].set_title("axes 4")

plt.tight_layout()
plt.show()
```

### Axis 객체와 축
- Axis 객체는 플롯의 가로축이나 세로축을 나타내는 객체
- https://matplotlib.org/api/axis_api.html
- twinx 명령: x 축을 공유하는 새로운 Axes 객체 생성

```python
fig, ax0 = plt.subplots()
ax1 = ax0.twinx()
ax0.set_title("2개의 y축 한 figure에서 사용하기")
ax0.plot([10, 5, 2, 9, 7], 'r-', label="y0")
ax0.set_ylabel("y0")
ax0.grid(False)
ax1.plot([100, 200, 220, 180, 120], 'g:', label="y1")
ax1.set_ylabel("y1")
ax1.grid(False)
ax0.set_xlabel("공유되는 x축")
plt.show()
```

### 바 차트(bar chart)
- bar: 세로 방향으로 바 차트
- barh: 가로 방향으로 바 차트
- http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.bar
- http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.barh

```python
import matplotlib as mpl
import matplotlib.pylab as plt

y = [2, 3, 1]
x = np.arange(len(y))
xlabel = ['가', '나', '다']
plt.title("Bar Chart")
plt.bar(x, y)
plt.xticks(x, xlabel)
plt.yticks(sorted(y))
plt.xlabel("가나다")
plt.ylabel("빈도 수")
plt.show()
```

#### 에러 바(error bar)
- xerr 인수나 yerr 인수를 지정
- alpha는 투명도를 지정(0이면 완전 투명, 1이면 완전 불투명)

```python
np.random.seed(0)

people = ['몽룡', '춘향', '방자', '향단']
y_pos = np.arange(len(people))
performance = 3 + 10 * np.random.rand(len(people))
error = np.random.rand(len(people))

plt.title("Barh Chart")
plt.barh(y_pos, performance, xerr=error, alpha=0.5)
plt.yticks(y_pos, people)
plt.xlabel('x 라벨')
plt.show()
```

### 파이 차트(pie chart)
- 카테고리 별 값의 상대적인 비교

### 컨투어 플롯(contour plot)
- 3차원 자료(입력 변수가 x, y 두 개이고 출력 변수가 z 하나인 경우)를 시각화하는 방법으로 명암이 아닌 등고선(contour)을 사용하는 방법
- contour 명령: 등고선만 표시
- contourf 명령: 색깔로 표시
- meshgrid 명령으로 x, y 그리드 포인트 행렬을 만들어야 한다.
- http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.contour
- http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.contourf

```python
def f(x, y):
    return (1 - x / 2 + x ** 5 + y ** 3) * np.exp(-x ** 2 - y ** 2)

n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
XX, YY = np.meshgrid(x, y)
ZZ = f(XX, YY)

plt.title("Contour plots")
plt.contourf(XX, YY, ZZ, alpha=.75, cmap='jet')
plt.contour(XX, YY, ZZ, colors='black')
plt.show()
```

### 3D 서피스 플롯
- Axes3D라는 3차원 전용 axes를 생성하여 입체적으로 표시
- plot_wireframe, plot_surface 명령
- http://matplotlib.org/mpl_toolkits/mplot3d/api.html#mpl_toolkits.mplot3d.axes3d.Axes3D.plot_wireframe
- http://matplotlib.org/mpl_toolkits/mplot3d/api.html#mpl_toolkits.mplot3d.axes3d.Axes3D.plot_surface

```python
from mpl_toolkits.mplot3d import Axes3D
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
XX, YY = np.meshgrid(X, Y)
RR = np.sqrt(XX**2 + YY**2)
ZZ = np.sin(RR)

fig = plt.figure()
ax = Axes3D(fig)
ax.set_title("3D Surface Plot")
ax.plot_surface(XX, YY, ZZ, rstride=1, cstride=1, cmap='hot')
plt.show()
```
