[Apriori](https://zephyrus1111.tistory.com/119)

```python
import numpy as np 
import pandas as pd 
```

```python
data = pd.read_csv('Market_Basket.csv', header = None) 
```

```python
data.head()
```

```python
transactions = []
```

```python
for i in range(data.shape[0]):
    transactions.append([str(data[j][i]) 
                         for j in range(data.shape[1]-data.isnull().sum(axis=1)[i])])
```

```python
transactions
```

# 2. apriori 모델 적용
`apriori(transaction_data, min_support, min_confidence, min_lift, min_length)
`
- min_length: item 갯수

```python
#!pip install apyori
```

```python
from apyori import apriori
```

```python
rules = apriori(transactions, min_support = 0.015, min_confidence = 0.2,  
                min_lift = 1, min_length = 2)
```

```python
results = list(rules) 
```

```python
results
```

```python
print(len(results))
```

```python
df=pd.DataFrame(results)
```

```python
df
```

```python
df.to_csv('apriori_result.csv')
```

```python
print(df.iloc[70:78][['items','support']]) 
```

# 3. 연관품목의 시각화

```python
ar=(df.iloc[1:74]['items'])    # item 2개인 것까지
```

```python
ar
```

```python
import networkx as nx
import matplotlib.pyplot as plt

plt.style.use('ggplot')
plt.figure(figsize=(9,6))

G = nx.Graph()
G.add_edges_from(ar)
pos = nx.spring_layout(G)

#nx.draw(G, pos, font_size=16, with_labels=False, edge_color='green',
#        node_size=800, node_color=['red','green','blue','cyan','orange','magenta'])
nx.draw(G, pos, font_size=16, with_labels=False, edge_color='green',
        node_size=500)

for p in pos:  
    pos[p][1] += 0.07
nx.draw_networkx_labels(G, pos)
plt.show()
```
